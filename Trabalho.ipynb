```python
data_hora = datetime.datetime.now()
url = 'http://brewbrothers.acesbyte.com.br/taps'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

forecast_item = soup.find_all(class_="row")

#for item in forecast_item: 
#    print(item.find('h2').text)
#    print(item.find('h4').text)
#    print(item.find(class_="col-xs-6").get_text())
#    print(item.find(class_="col-xs-6 text-right").get_text())
```

```python
for item in forecast_item: 
    nome = item.find('h2').text
    desc = item.find('h4').text
    print(nome)
    print(desc)
    for preco in item.find_all(class_="col-xs-6"):
        valor = preco.text
        print(valor)
               
```

```python
client = MongoClient('localhost', 27017)

for item in forecast_item: 
    print(item.find('h2').text)
    print(item.find('h4').text)
    for preco in item.find_all(class_="col-xs-6"):
        print(preco.text)

for item in forecast_item: 
    post = {
        "data" : data_hora,
        "nome" : item.find('h2').text,
        "desc" : item.find('h4').text,
        "tamanho" : item.find(class_="col-xs-6").get_text(),
        "valor" : item.find(class_="col-xs-6 text-right").get_text(),
    }
    post_id = client.TBB.taps.insert_one(post).inserted_id

client.close
```
